```python
import pandas as pd
import matplotlib.pyplot as plt

# CSV dosyasını oku
df = pd.read_csv("yolov5_outputs/runs/train/yolov5_sirst_augmented3/results.csv")
df.columns = df.columns.str.strip()  # Baş ve sondaki boşlukları kaldır

# Epoch numaraları
epochs = df.index.values

# 1️⃣ Box Loss Grafiği (Train + Val)
plt.figure(figsize=(8, 5))
plt.plot(epochs, df['train/box_loss'], label='Train Box Loss')
plt.plot(epochs, df['val/box_loss'], label='Val Box Loss')
plt.xlabel('Epoch')
plt.ylabel('Box Loss')
plt.title('Box Loss vs Epoch')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig("yolo_report_images/box_loss_plot.png")
plt.show()

# 2️⃣ Objectness Loss Grafiği (Train + Val)
plt.figure(figsize=(8, 5))
plt.plot(epochs, df['train/obj_loss'], label='Train Obj Loss')
plt.plot(epochs, df['val/obj_loss'], label='Val Obj Loss')
plt.xlabel('Epoch')
plt.ylabel('Objectness Loss')
plt.title('Objectness Loss vs Epoch')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig("yolo_report_images/obj_loss_plot.png")
plt.show()

# 3️⃣ mAP@0.5 Grafiği
plt.figure(figsize=(8, 5))
plt.plot(epochs, df['metrics/mAP_0.5'], label='mAP@0.5', color='darkorange')
plt.xlabel('Epoch')
plt.ylabel('mAP@0.5')
plt.title('mAP@0.5 vs Epoch')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig("yolo_report_images/map_0.5_plot.png")
plt.show()

```
